```python
df_p['unidad_sat_id'] = df_p['unidad_sat'].map({
    'KGM': 1,
    'LTR': 2,
    'H87': 3,
})
```

```python
cat_map = {
    'INSECTICIDAS': '2679',
    'FUNGICIDAS': '2680',
    'NUTRIENTES FOLIARES': '2699',
    'HERBICIDAS': '2725',
    'BIOESTIMULANTES': '2726',
    'FERTILIZANTES SOLUBLES': '2727',
    'COADYUVANTES': '2728',
    'NUTRIENTES PARA FERTIRIGACIÓN': '2729',
    'FERTILIZANTES GRANULADOS': '2730',
    'MEJORADORES DE SUELO': '2731',
    'ACARICIDAS': '2732',
    'BACTERICIDAS': '2733',
    'BIOPROTECTORES': '2734',
    'REGULADOR DE CRECIMIENTO': '2735',
    'ENRAIZADOR': '2736',
    'RODENTICIDA': '2737',
    '(Ninguna)' : '2738'
}
cat_map
```

```python
df_p['cat_id'] = df_p['cat'].map(cat_map)
```

```python
df_p['clave_sat'] = df_p['clave_sat'].fillna(0).astype(int).astype(str)
df_p['cat_id'] = df_p['cat_id'].fillna(0).astype(int).astype(str)
df_p['created_by'] = 1
df_p['created_at'] = int(time.time())
```

```python
df_p.head(2)
```

```python
#inser 
cursor = conn.cursor()
insert_query = """
INSERT INTO producto (clave, clave_sat, nombre, ieps, iva, unidad_medida_id, categoria_id,
 precio_publico, precio_sub, precio_mayoreo, created_by, created_at)
 values (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""
data_to_insert = df_p[['codigo', 'clave_sat', 'nombre', 'IEPS', 'IVA', 'unidad_sat_id', 'cat_id',
                       'p_publico', 'p_sub', 'p_mayoreo', 'created_by', 'created_at']].values.tolist()

cursor.executemany(insert_query, data_to_insert)
conn.commit()
cursor.close()
```

## PROVEEDOR

```python
df_pv = pd.read_excel(path_proveedores)
```

```python
df_pv.head(3)
#df_pv['País'].unique()
```

```python
columns_ocupar = ['code', 'nombre', 'rfc']
df_pv = df_pv[columns_ocupar]
df_pv['created_by'] = 1
df_pv['created_at'] = int(time.time())
df_pv['rfc'] = df_pv['rfc'].fillna("") .astype(str)
df_pv['nombre'] = df_pv['nombre'].apply(lambda x: x.strip().upper() if isinstance(x, str) else x)
df_pv['nombre'] = df_pv['nombre'].fillna("") .astype(str)
df_pv['nombre_'] = df_pv['nombre'].fillna("") .astype(str)

df_pv['code'] = df_pv['code'].fillna("") .astype(str)


df_pv
```

```python
#inser 
cursor = conn.cursor()
insert_query = """
INSERT INTO proveedor (nombre, razon_social, rfc, created_by, created_at)
 values (%s, %s, %s, %s, %s)
"""
data_to_insert = df_pv[['nombre', 'nombre_', 'rfc', 
                        'created_by', 'created_at']].values.tolist()

cursor.executemany(insert_query, data_to_insert)
conn.commit()
cursor.close()
```
